##how to analyse data frame 
.head()-->it shows the first N rows in data,BY default 5
.shape--> it shows the total now of rows and columns in a dataframe
.index()--> this attributes provide the index of the dataframe
.column---> show the name of each column
.dtype--> shows the data type of each column
.unique--> it shows unique vales of each column it is applied to only one column 
.nunique-->shows total no of unique values in the each column applied to total dataframe
.count--> it shows the count of all non null values in the column , it can be used to single column or the complete dataframe
.value_count--> it shows all the unique value in the column and its count, applicable only for an single column
.info--> provide basic infomation
